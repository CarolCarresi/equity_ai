```python
from textblob import TextBlob

def get_sentiment_score(text):
    # Create a TextBlob object
    blob = TextBlob(text)

    # Get the sentiment polarity (range -1 to 1)
    sentiment_score = blob.sentiment.polarity

    return sentiment_score
```

```python
scores = []
for n in news:
    d = []
    for text in n:
        s = get_sentiment_score(text)
        d.append(s)
    min_val = min(d)
    max_val = max(d)
    avg_val = sum(d)/len(d)
    median_val = sorted(d)[len(d)//2]
    scores.append([min_val, max_val, avg_val, median_val])
```

```python
scores
```

```python
import numpy as np

X = np.array(scores)
y = df['label'].values
```

```python
# perform LOO cross validation using logistic regression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score

loo = LeaveOneOut()
loo.get_n_splits(X)
feat_importances = []
mean_acc = []
for i in range(1, 10):
    ys_test = []
    ys_pred = []
    for train_index, test_index in loo.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        clf = DecisionTreeClassifier(random_state=i, max_depth=5).fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        ys_test.append(y_test[0])
        ys_pred.append(y_pred[0])
        feat_importances.append(clf.feature_importances_)

    mean_acc.append(accuracy_score(ys_test, ys_pred))

print(np.mean(mean_acc))
feat_importances = np.mean(feat_importances, axis=0)
```

```python
#plot feature importances
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x=feat_importances, y=['min', 'max', 'avg', 'median'])
```
